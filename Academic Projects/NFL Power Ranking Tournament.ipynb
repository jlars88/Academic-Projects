```python
#The Tournament  Algorithm

def tourney(g, n):
  Dictionary2 = {vertex: 0 for vertex in g}
  for vertex in g:
    for adjacent in g[vertex]:
      Dictionary2[vertex] = Dictionary2[vertex] + 1
  n = n-1

  while n> 0:
    Dictionary3 = {vertex: 0 for vertex in g}

    for vertex in g:
      for adjacent in g[vertex]:
        Dictionary3[vertex] = Dictionary3[vertex]+ Dictionary2[adjacent]

    Dictionary2 = Dictionary3
    n = n-1

  return Dictionary2
```

```python
# Smaller test graphs for the Algorithm
g = {"A": {"B", "C"},
     "B": {"D"},
     "C": {"B"},
     "D": {"A", "C"}}

best = tourney(g, 1)
sorted_results = dict(sorted(best.items(), key=lambda item: item[1], reverse=True))
print(sorted_results)
best = tourney(g, 5)
sorted_results = dict(sorted(best.items(), key=lambda item: item[1], reverse=True))
print(sorted_results)
```
