```python
import pandas as pd
# import numpy as np
import seaborn as sns
qb = pd.read_excel("../Python_data/qbgradeTEST1.xlsx")
qb = qb.rename(columns={"CORRECT PRESNAP ALLIGNMENT": "CORRECT PRESNAP ALIGNMENT"})
qb = qb.rename(columns={"SIGHT ADJUST": "LEVERAGE"})
qb
```

```python
# Creating the PRESNAP Grade
presnap_cols = ['CORRECT POINT', 'CORRECT PRESNAP ALIGNMENT', 'SET PROTECTION', 'FIRST LOOK']

qb[presnap_cols] = qb[presnap_cols].apply(pd.to_numeric, errors='coerce')
qb[presnap_cols] = qb[presnap_cols].fillna(qb[presnap_cols].mean())
qb["PRESNAP_GRADE"] = qb[presnap_cols].sum(axis=1)
```

```python
# Creating the PASSING_GRADE
passing_cols = ['THROW DIFFICULTY', 'THROW PLACEMENT', 'THROW TIMING', 'TURNOVER WORTHY']

qb[passing_cols] = qb[passing_cols].apply(pd.to_numeric, errors='coerce')
qb[passing_cols] = qb[passing_cols].fillna(qb[passing_cols].mean())
qb["PASSING_GRADE"] = qb[passing_cols].mean(axis=1)

```

```python
# Creating the ATHLETIC Grade
athletic_cols = ['ELUSIVENESS', 'ARM ANGLE', 'EYES DOWNFIELD', 'ESCAPABILITY']

qb[athletic_cols] = qb[athletic_cols].apply(pd.to_numeric, errors='coerce')
qb[athletic_cols] = qb[athletic_cols].fillna(qb[athletic_cols].mean())
qb["ATHLETIC_GRADE"] = qb[athletic_cols].sum(axis=1)
```

```python
# Creating the POSTSNAP Grade
postsnap_cols = ['CORRECT READ', 'CORRECT SIDE', 'PROGRESSION UNDER PRESSURE', 'LEVERAGE']

qb[postsnap_cols] = qb[postsnap_cols].apply(pd.to_numeric, errors='coerce')
qb[postsnap_cols] = qb[postsnap_cols].fillna(qb[postsnap_cols].mean())
qb["POSTSNAP_GRADE"] = qb[postsnap_cols].sum(axis=1)
```

```python
# Creating the MECHANICS Grade
mechanics_cols = ['DROPBACK', 'ALIGNMENT', 'BALL SECURITY', 'LAUNCHPOINT']

qb[mechanics_cols] = qb[mechanics_cols].apply(pd.to_numeric, errors='coerce')
qb[mechanics_cols] = qb[mechanics_cols].fillna(qb[mechanics_cols].mean())
qb["MECHANICS_GRADE"] = qb[mechanics_cols].sum(axis=1)
```
