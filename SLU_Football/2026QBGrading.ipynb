```python
import pandas as pd
# import numpy as np
import seaborn as sns
qb = pd.read_excel("../Python_data/qbgradeTEST1.xlsx")
qb
```

```python
# Creating PASSING_GRADE
passing_cols = ['THROW DIFFICULTY', 'THROW PLACEMENT', 'THROW TIMING', 'TURNOVER WORTHY']

qb[passing_cols] = qb[passing_cols].apply(pd.to_numeric, errors='coerce')
qb[passing_cols] = qb[passing_cols].fillna(qb[passing_cols].mean())
qb["PASSING_GRADE"] = qb[passing_cols].mean(axis=1)


qb  # now each row (play) has PASSING_GRADE
```
